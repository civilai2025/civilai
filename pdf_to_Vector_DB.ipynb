```python
# 📦 Imports
!pip install numpy==1.23.5
import os, glob, gc
from tqdm import tqdm
from PyPDF2 import PdfReader
from pdf2image import convert_from_path
import numpy as np

from paddleocr import PaddleOCR
from langdetect import detect
from transformers import MarianMTModel, MarianTokenizer
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.schema import Document as LCDocument

# 🌐 Translation Model (Hindi to English)
model_name = 'Helsinki-NLP/opus-mt-hi-en'
tokenizer = MarianTokenizer.from_pretrained(model_name)
translator = MarianMTModel.from_pretrained(model_name)

def translate_hi_to_en(text, max_length=512):
    sentences = text.split("\n")
    translated = []
    for sent in sentences:
        if not sent.strip():
            translated.append("")
            continue
        inputs = tokenizer(sent, return_tensors="pt", padding=True, truncation=True, max_length=max_length)
        translated_tokens = translator.generate(**inputs)
        english = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
        translated.append(english)
    return "\n".join(translated)

# 📥 Get PDF file paths
def get_all_documents_from_folder(folder_path):
    return glob.glob(os.path.join(folder_path, '**', '*.pdf'), recursive=True)

# ⚡ Fast OCR using PaddleOCR
ocr_engine = PaddleOCR(lang='en', use_angle_cls=True, use_gpu=False)

def fast_ocr(pdf_path):
    pages = convert_from_path(pdf_path, dpi=150)  # lower dpi for speed
    full_text = ''
    for page in pages:
        result = ocr_engine.ocr(np.array(page), cls=True)
        if result and result[0]:  # Check if OCR detected any text lines
            page_text = '\n'.join([line[1][0] for line in result[0]])
        else:
            page_text = ''
        lang = detect(page_text) if page_text.strip() else 'en'
        if lang == 'hi':
            page_text = translate_hi_to_en(page_text)
        full_text += page_text + '\n'
    return full_text


# def fast_ocr(pdf_path):
#     pages = convert_from_path(pdf_path, dpi=150)  # lower dpi for speed
#     full_text = ''
#     for page in pages:
#         result = ocr_engine.ocr(np.array(page), cls=True)
#         page_text = '\n'.join([line[1][0] for line in result[0]])
#         lang = detect(page_text) if page_text.strip() else 'en'
#         if lang == 'hi':
#             page_text = translate_hi_to_en(page_text)
#         full_text += page_text + '\n'
#     return full_text

# 🧠 Text Extractor with fallback to OCR
def extract_text(file_path):
    if file_path.endswith('.pdf'):
        try:
            reader = PdfReader(file_path)
            raw_text = ''.join((p.extract_text() or '') + "\n" for p in reader.pages)
            if raw_text.strip():
                return raw_text
            return fast_ocr(file_path)
        except:
            return fast_ocr(file_path)
    return ''
```

```python
from google.colab import drive
drive.mount('/content/drive')

folder_path = "/content/drive/MyDrive/model_and_dataset/NH-458"
db_path = "/content/drive/MyDrive/model_and_dataset/Untitled folder/all_text_NH_458.docx"

# Load files
doc_files = get_all_documents_from_folder(folder_path)
all_chunks = []

# Split into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

for file in tqdm(doc_files, desc="Processing PDFs"):
    text = extract_text(file)
    chunks = splitter.split_text(text)
    for c in chunks:
        if c.strip():  # ✅ No min length condition
            all_chunks.append(LCDocument(page_content=c, metadata={"source": os.path.basename(file)}))

# Embed & store
embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
if all_chunks:
    vector_db = FAISS.from_documents(all_chunks, embedding_model)
    vector_db.save_local(db_path)
    print(f"✅ Vector store saved at {db_path}")
else:
    print("⚠️ No valid chunks found. Nothing to save.")
```

```python
from langchain.vectorstores import FAISS
from sentence_transformers import CrossEncoder
from transformers import pipeline as hf_pipeline
from langchain.llms import HuggingFacePipeline
from langchain.embeddings import HuggingFaceEmbeddings

# ✅ Load embedding model and FAISS vector DB
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorDB = FAISS.load_local("/content/drive/MyDrive/VDb", embedding_model, allow_dangerous_deserialization=True)

# ✅ Load cross-encoder reranker
reranker = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')

# ✅ HuggingFace LLM for QA (text2text generation)
qa_pipe = hf_pipeline("text2text-generation", model="google/flan-t5-base", max_length=256)  # Limit output size
llm = HuggingFacePipeline(pipeline=qa_pipe)

# ✅ Ask your query
query = "As per Para-5.0 of MoRT&H Circular Dated 04th June 2024, clear deck width of bridges/grade"

# Step 1: Initial search from Vector DB
top_k = 10
docs = vectorDB.similarity_search(query, k=top_k)

# Step 2: Rerank using CrossEncoder
pairs = [(query, doc.page_content) for doc in docs]
scores = reranker.predict(pairs)

# Step 3: Sort and get top 3 reranked
top_docs = [doc for _, doc in sorted(zip(scores, docs), key=lambda x: x[0], reverse=True)][:3]

# Step 4: Truncate context to stay within token limits (basic way)
combined_context = "\n\n".join(doc.page_content[:1000] for doc in top_docs)  # Keep only first 1000 chars per doc

# Step 5: Run QA directly using LLM
prompt = f"Answer the question based on the context:\n\n{combined_context}\n\nQuestion: {query}"
response = llm.invoke(prompt)  # ← use .invoke() to avoid deprecation warnings

# Step 6: Output
print("🔍 Answer:\n", response)
print("\n📁 Sources:")
for i, doc in enumerate(top_docs, 1):
    source = doc.metadata.get("source", "Unknown source")
    print(f"{i}. {source}")
```
